```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data16 = np.loadtxt("dev16/data_not100.txt", delimiter=",")
data18 = np.loadtxt("dev18/data_not100.txt", delimiter=",")
print(np.shape(data18))

n16 = 2*data16[:,0]**3 - data16[:,0]**2
t16 = n16/data16[:,1]
plt.plot(n16, t16/1e9)
plt.xscale("log")
plt.xlabel("Matrix Size")
plt.ylabel("Gflop/s")
plt.title("Gflops/s for HPCC intel16 nodes")
plt.savefig("dev16/speed_plot.png")
```

```python
n18 = 2*data18[:,0]**3 - data18[:,0]**2
t18 = n18/data18[:,1]
plt.plot(n18, t18/1e9)
plt.xscale("log")
plt.xlabel("Matrix Size")
plt.ylabel("Gflop/s")
plt.title("Gflops/s for HPCC intel18 nodes")
plt.savefig("dev18/speed_plot.png")
```
