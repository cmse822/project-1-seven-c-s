```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
data = np.loadtxt('data_cache_info.txt',delimiter=',')
```

```python
matrix_size = data[:,0]
gflops = data[:,3]
```

```python
plt.style.use('ggplot')
plt.scatter(matrix_size,gflops,s=5)
plt.title('GFlops vs Matrix Size; M1 Macbook Air')
plt.ylabel('Giga-Flops')
plt.xlabel('Matrix Size')
```
